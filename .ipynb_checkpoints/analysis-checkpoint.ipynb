````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
````

````python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
````

````python
df = pd.read_excel('https://data.mendeley.com/datasets/jcj9j7b4bd/1/files/9b6811dd-6068-4c80-93d1-bd4cea1c4fe6/Data%20-%20Video%20gaming%20and%20working%20memory.xlsx?dl=1')
````

````python
df.isnull().values.any()
````

````python
df.isnull().sum()
````

````python
df.Videogaming_hoursperweek.describe()
````

````python
df.head()
````

````python
df.describe()
````

````python
df['gaming hours'] = pd.cut(df['Videogaming_hoursperweek'], [0, 1, 5, 10, 20, 71], labels=['0','1-5','5-10','10-20','20+'])
````

````python
df['gaming hours'] = df['gaming hours'].fillna('0')
````

````python
grouped_by_mean = df.groupby('gaming hours').mean().reset_index()
````

````python
grouped_by_mean
````

````python
df.groupby('gaming hours')['Nback_WM_Composite_n503_AverageZ'].mean().plot()
````

````python
df['gamer'] = np.where(df['Videogaming_hoursperweek']>0, 'yes', 'no')

df
````

````python
df.head()
````

````python
sns.relplot(x='Age_in_years',y='Videogaming_hoursperweek', hue='Education_highestdegree', data=df)
````

````python
ax = sns.boxplot(x='gaming hours', y='Nback_WM_Composite_n503_AverageZ', data=df)
plt.xlabel('Hours spent gaming per week')
plt.ylabel('N-back Task Score')
plt.savefig('nback.png', dpi=100)
````

````python
ax = sns.boxplot(x='gaming hours', y='Spatial_WM_Composite_n503_AverageZ', data=df)
plt.xlabel('Hours spent gaming per week')
plt.ylabel('Spatial Working Memory Score')
plt.savefig('spatial.png', dpi=100)
````

````python
ax = sns.boxplot(x='gaming hours', y='Verbal_WM_Composite_n503_AverageZ', data=df)
plt.xlabel('Hours spent gaming per week')
plt.ylabel('Verbal Working Memory Score')
plt.savefig('verbal.png', dpi=100)
````

````python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="Videogaming_hoursperweek", y="Nback_WM_Composite_n503_AverageZ",
                      hue="gamer",
                         palette="Set2",
                         data=df)
````

````python
g = sns.catplot(x='gamer', y='Nback_WM_Composite_n503_AverageZ', kind='bar',
                height=4, aspect=.7, labels=['gamer','non gamer'], data=df)

````

````python
g = sns.catplot(x='gamer', y='Spatial_WM_Composite_n503_AverageZ', kind='bar',
                height=4, aspect=.7, data=df, legend=True)
plt.ylabel('Spatial Memory Score')
plt.xlabel('Gamer')


plt.savefig('spatial-gamer.png', dpi=100)
````

````python
g = sns.catplot(x='gamer', y='Nback_WM_Composite_n503_AverageZ', kind='bar',
                height=4, aspect=.7, data=df, legend=True)
plt.ylabel('N-back Task Score')
plt.xlabel('Gamer')


plt.savefig('n-back-gamer.png', dpi=100)
````
